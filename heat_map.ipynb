```python
def create_fasta(sequence, filename, description="sequence"):
    with open(filename, "w") as fasta_file:
        fasta_file.write(f">{description}\n")
        fasta_file.write(sequence + "\n")
```

```python
def build_chain(reference, new_chain_base):
    new_chain = []
    for ref, base in zip(reference, new_chain_base):
        if base == '.':
            new_chain.append(ref)
        else:
            new_chain.append(base)
    return ''.join(new_chain)
```

```python
fasta_files = os.listdir("./fasta_files")
sequences = {}

for fasta_file in fasta_files:
    with open(f"./fasta_files/{fasta_file}", "r") as fasta:
        sequences[fasta_file.split(".")[0]] = fasta.read().split("\n")[1]
```

```python
order = ["gro87", "senegal88", "india90", "donrep89", "china90", "malaysia90", "venezue90",
         "pueb91", "slp95", "phi56", "phi64", "phi84", "pric86", "indon77", "braz87",
         "caledo84", "salv83", "tait79", "tait85", "salv94", "mex84", "indon73", "tailan84", "tailan63"]

ordered_sequences = {key: sequences[key] for key in order if key in sequences}
reference_sequence = ordered_sequences["gro87"]
heatmap_data = []
nucleotide_colors = {'A': 1, 'T': 2, 'C': 3, 'G': 4}
color_labels = ['Blanco', 'Rojo (A)', 'Naranja (T)', 'Azul (C)', 'Verde (G)']
change_mask = [False] * len(reference_sequence)
for key, sequence in ordered_sequences.items():
    row = []
    for i, nucleotide in enumerate(sequence):
        if key == "gro87":
            continue 
        if nucleotide != reference_sequence[i]:
            change_mask[i] = True

row_gro87 = [
    nucleotide_colors.get(reference_sequence[i], 0) if change_mask[i] else 0
    for i in range(len(reference_sequence))
]
heatmap_data.append(row_gro87)

for key, sequence in ordered_sequences.items():
    if key == "gro87":
        continue
    row = []
    for i, nucleotide in enumerate(sequence):
        if nucleotide == reference_sequence[i]:
            row.append(0)
        else:
            row.append(nucleotide_colors.get(nucleotide, 0))
    heatmap_data.append(row)
```

```python
heatmap_array = np.array(heatmap_data)
cmap = ListedColormap(['white', 'red', 'orange', 'blue', 'green'])
bold_labels = ["gro87", "senegal88", "india90", "donrep89", "china90", "malaysia90", "venezue90", "pueb91", "slp95", "phi56"]

start = 901
end = 1182
step = 10
x_labels = list(range(start, end + 1, step))
x_ticks = list(range(0, len(reference_sequence), step))
yticks = ['gro87'] + [key for key in ordered_sequences.keys() if key != "gro87"]

plt.figure(figsize=(20, 8))
plt.imshow(heatmap_array, cmap=cmap, aspect="auto")
cbar = plt.colorbar(ticks=range(1, 5))
cbar.ax.set_yticklabels(['A', 'T', 'C', 'G'])
for idx, label in enumerate(yticks):
    if label in bold_labels:
        plt.text(-0.5, idx, label, fontsize=10, fontweight='bold', va='center', ha='right')
    else:
        plt.text(-0.5, idx, label, fontsize=10, va='center', ha='right')

plt.yticks([], [])
plt.xticks(ticks=x_ticks, labels=x_labels)
plt.grid(color='gray', linestyle='--', linewidth=0.5, which='both', axis='both')
plt.xlabel("PosiciÃ³n en la secuencia")
plt.title("Modificaciones de las secuencias respecto a la secuencia de referencia (gro87)")
plt.show()
```
